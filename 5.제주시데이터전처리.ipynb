```python
df = pd.read_excel('dataset\processedData\읍면동별_데이터.XLSX', engine='openpyxl')
df = df.drop(['제주 인구(남) (명)', '제주 인구(여) (명)',
       '한국인 세대수 (세대)', '한국인(계) (명)', '한국인(남) (명)', '한국인(여) (명)', '외국인(계) (명)',
       '외국인(남) (명)', '외국인(여) (명)'], axis=1)

# 행정시별 공백제거
pre_data = []
for i in range(len(df)):
    temp = df['행정시별(2)'][i].replace(" ", "")
    pre_data.append(temp)
df['행정시별(2)'] = pre_data
```

```python
df = pd.read_csv('dataset\제주특별자치도_연령대별13세이상인구의료서비스이용률_2020.csv')
df['시도'] = '제주'
df = df.dropna(axis=0)
df = df.groupby('연령대').mean().drop('기준 연도', axis=1)
df = df.reset_index()
```

```python
df_10_to_60per = df[0:6]['이용률(%)'].mean()
df_65per = df[df['연령대'] == '65세이상']['이용률(%)']
```

```python
weight_65_per = float((df_65per / df_10_to_60per))
print("타 연령대비 65세 이상의 병원 이용률 : ", weight_65_per)
```
